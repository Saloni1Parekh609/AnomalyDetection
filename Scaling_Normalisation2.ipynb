<h1>3. Random All SRI</h1>

```python
df_3 = df_scaled
df_3.columns = df_scaled.columns
df_3.head()
```

```python
missing_column = df_3.columns[df_3.isna().any()].tolist()
```

```python
def random_imputation(dft, feature):
    number_missing = dft[feature].isnull().sum()
    observed_value = dft.loc[dft[feature].notnull(), feature]
    dft.loc[dft[feature].isnull(), feature + '_imp'] = np.random.choice(observed_value, number_missing, replace = True)
    return dft
```

```python
for feature in missing_column:
    df_3[feature + '_imp'] = df_3[feature]
    df_3 = random_imputation(df_3, feature)
```

```python
#Stochastic Regression Imputation
from sklearn import linear_model
random_data = pd.DataFrame(columns = ["Ran" + name for name in missing_column])

for feature in missing_column:
    #print('feature')
    yx = np.nan_to_num(df_3[feature + '_imp'])
    random_data["Ran" + feature] = df_3[feature + '_imp']
    parameters = list(set(df_3.columns) - set(missing_column) - {feature + '_imp'})
    t = np.nan_to_num(df_3[parameters])
    
    model = linear_model.LinearRegression()
    model.fit(X = t, y = yx)
    
    predict = model.predict(t)
    std_error = (predict[df_3[feature].notnull()] - df_3.loc[df[feature].notnull(), feature + '_imp']).std()
    
    random_predict = np.random.normal(size = df_3[feature].shape[0], 
                                      loc = predict, 
                                      scale = std_error)
    random_data.loc[(df_3[feature].isnull()) & (random_predict > 0), "Ran" + feature] = random_predict[(df_3[feature].isnull()) & 
                                                                            (random_predict > 0)]
```

```python
df_4 = pd.DataFrame(random_data, columns = random_data.columns)
df_4.head()
```
