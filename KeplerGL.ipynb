```python
from geopy.geocoders import Nominatim
#geopy.geocoders.options.default_user_agent = "my-application"

geolocator = Nominatim(user_agent= "myGeocoder")
def geolocate(country):
    try:
        loc = geolocator.geocode(country)
        return (loc.latitude, loc.longitude)
    except:
        return [0,0]
```

```python
lat = []
lng = []
for i in range(df_2013.shape[0]):
    print(df_2013.iloc[i,1])
    geocode = geolocate(df_2013.iloc[i,1])
    lat.append(geocode[0])
    lng.append(geocode[1])
    
#df_2003['Latitude'] = lat
#df_2003['Longitude'] = lng
#df_2003.head()
```

```python
df_2013 = pd.DataFrame()
df_2013['Latitude'] = lat
df_2013['Longitude'] = lng
df_2013.head()
```

```python
from keplergl import KeplerGl
import geopandas as gpd

map_1 = KeplerGl(height=500)
#df=pd.read_csv('df2001.csv')
worldmap['Longitude']=pd.to_numeric(worldmap['Longitude'])
worldmap['Latitude']=pd.to_numeric(worldmap['Latitude'])
gdf = gpd.GeoDataFrame(worldmap, geometry=gpd.points_from_xy(worldmap.Longitude, worldmap.Latitude))
map_1.add_data(data=gdf, name = 'd1')
worldmap_pred['Longitude']=pd.to_numeric(worldmap_pred['Longitude'])
worldmap_pred['Latitude']=pd.to_numeric(worldmap_pred['Latitude'])
gdf = gpd.GeoDataFrame(worldmap_pred, geometry=gpd.points_from_xy(worldmap_pred.Longitude, worldmap_pred.Latitude))
map_1.add_data(data=gdf, name = 'd2')
map_1.save_to_html(file_name='w2.html')
```
